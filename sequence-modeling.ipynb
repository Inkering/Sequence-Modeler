```python
test_text = df['transcript'][0]; #uncomment to see full transcript... these are long strings
```

lets turn the data into tokens which we can then analyze the frequency of later -> for N-grams sequence learning predictions

```python
tokens = re.split('[^A-Za-z]+', test_text.lower())
# may need more filtering here, also expand beyond one transcript to include
# all transcripts
```

Build our frequency dictionary, for a 5 grams setup

```python
gram_list = dict()
for i in range(len(tokens)-1):
    test = tuple(tokens[i:i+2]) 
    if test in gram_list:
        gram_list[test] += 1
    else:
        gram_list[test] = 1
gram_list = [(k, v) for k, v in gram_list.items()] 
gram_list.sort(key = lambda x: x[1])
gram_list.reverse()
```

Sort to look at most common ones first

```python
gram_list;
```

```python
def choicer(choices):
    total_options = sum(w2 for w1, w2 in choices)
    r = random.uniform(0, total_options)
    counter = 0
    for w1, w2 in choices:
        if counter + w2 > r:
            return w1
        counter += w2
```

```python
new_word = "good"
for i in range(30): # 10 word sentence
    print(new_word, end=" ")
    choices = [item for item in gram_list if item[0][0] == new_word]
    if not choices:
        break
    new_word = choicer(choices)[1] # grab the next word in the seq
```
